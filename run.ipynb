# Per-participant EEG sentiment model
Run the cells in order. The first one sets everything up; the second asks for your data folders and where to save.

```python
# Setup: pulls the project, installs deps, mounts Drive. Safe to re-run.
import os, sys, subprocess

REPO_URL = "https://github.com/parmisbathaeiyan/teco-eeg-nlp.git"
REPO_DIR = "teco-eeg-nlp"

if not os.path.isdir("src"):
    if not os.path.isdir(REPO_DIR):
        subprocess.run(["git", "clone", "-q", REPO_URL], check=True)
    os.chdir(REPO_DIR)

subprocess.run(["pip", "install", "-q", "-r", "requirements.txt"], check=False)
sys.path.insert(0, "src")

try:
    from google.colab import drive
    drive.mount("/content/drive")
except ModuleNotFoundError:
    pass

print("ready - working dir:", os.getcwd())
```

```python
from per_participant_eeg import run

trt_dir    = input("TRT_Total folder (the *_trt_total.pickle files): ").strip()
labels_dir = input("Sentiment Excel folder (Total_Sentiment_Raw_*.xlsx): ").strip()
out_dir    = input("Save results to [results]: ").strip() or "results"

results = run(trt_dir, labels_dir, out_dir)
```
